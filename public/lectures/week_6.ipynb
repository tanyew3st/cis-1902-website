## Array Manipulation in NumPy

Manipulating the shape and size of arrays is a crucial aspect of NumPy.

### Topics Covered

- `shape` and `reshape`: Understanding and modifying array dimensions.
- `max`, `min`, `argmax`, `argmin`: Finding maximum and minimum values and their indices.
- Slicing and Indexing: Accessing and modifying array elements.
- Broadcasting: How NumPy handles operations on arrays of different shapes.


```python
# Creating an array and reshaping
original_array = np.arange(1, 7)  # A 1D array
reshaped_array = original_array.reshape((2, 3))  # Reshaped to a 2D array

two_d_array = np.array([[1, 2, 3], [4, 5, 6]])

# Output
print("Original Array:", original_array)
print("Reshaped Array:", reshaped_array)
print("2D Array:", two_d_array)
print("Array shapes:", original_array.shape, reshaped_array.shape, two_d_array.shape)
```

```python
# Appending, deleting, inserting
original_array = np.array([1, 2, 3])

# Appending
appended_array = np.append(original_array, 4)
print("Appended Array:", appended_array)

# Deleting
deleted_array = np.delete(original_array, 1)
print("Deleted Array:", deleted_array)

# Inserting
inserted_array = np.insert(original_array, 1, 5)
print("Inserted Array:", inserted_array)

# Everything is immutable
print("Original Array:", original_array)
```

```python
arr = np.arange(1, 7)

# Using max, min, argmax, and argmin
max_value = arr.max()
min_value = arr.min()
max_index = arr.argmax()
min_index = arr.argmin()

# Output
print("Max Value:", max_value)
print("Min Value:", min_value)
print("Index of Max:", max_index)
print("Index of Min:", min_index)
```

```python
reshaped_array = np.arange(1, 7).reshape((2, 3))

# Array slicing and indexing
sliced_array = reshaped_array[:, 1:3]  # Slicing columns
sliced_array[0, 1] = 100

# Demonstrating broadcasting and slices being pointers
print("Modified Reshaped Array:\n", reshaped_array)
print("Sliced Array:\n", sliced_array)
```

# Understanding NumPy vstack and hstack

NumPy offers various functions to manipulate the structure of arrays. Two such functions are `vstack` and `hstack`, which stack arrays vertically and horizontally, respectively.

## NumPy vstack (Vertical Stack)

- `vstack` stacks arrays in sequence vertically (row-wise).
- It is equivalent to concatenation along the first axis after 1-D arrays of shape `(N,)` have been reshaped to `(1,N)`.
- Useful for combining arrays with the same number of columns.

## NumPy hstack (Horizontal Stack)

- `hstack` stacks arrays in sequence horizontally (column-wise).
- It is equivalent to concatenation along the second axis, except for 1-D arrays where it concatenates along the first axis.
- Useful for combining arrays with the same number of rows.


```python
# Create two arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Vertically stack the arrays
v_stacked = np.vstack((a, b))

print("Vertical Stack Result:\n", v_stacked)
```

```python
# Create two arrays
a = np.array([[1], [2], [3]])
b = np.array([[4], [5], [6]])

# Horizontally stack the arrays
h_stacked = np.hstack((a, b))

print("Horizontal Stack Result:\n", h_stacked)
```

# Why is NumPy Faster? Vectorization and More

NumPy's performance advantage is due to several key factors:

## Vectorization

- **Vectorization** refers to implementing operations so they process multiple elements simultaneously rather than iterating over them individually.
- NumPy leverages vectorized operations that are implemented in low-level languages like C and Fortran.
- This approach is highly efficient, as it minimizes the loop overhead in Python and takes advantage of modern CPU capabilities.

## Hardware Acceleration

- NumPy can utilize features like SIMD (Single Instruction, Multiple Data) to perform parallel operations on modern CPUs.
- This enables processing multiple data points with a single instruction, greatly enhancing computation speed.

## Less Overhead

- Python lists are dynamic and can store elements of different types, adding overhead.
- NumPy arrays are homogenous and have a fixed type, reducing memory usage and computational overhead.

## Optimized Algorithms

- NumPy functions use highly optimized algorithms, often written in low-level languages, for mathematical operations.
- This results in faster execution compared to native Python operations.

Running speed tests between Python's native operations and NumPy's vectorized operations clearly demonstrates the significant performance improvements offered by NumPy, especially for large-scale data processing and complex computations.

```python
import random
import time

# Setting initial parameters
bet_amount = 100
initial_balance = 100000
win_probability = 0.49
num_rounds = 50000

# Start timing the standard Python implementation
start_time_python = time.time()

# Generate random outcomes based on win probability using standard Python
outcomes_python = [
    bet_amount if random.random() < win_probability else -bet_amount
    for _ in range(num_rounds)
]

# Calculate cumulative winnings to get the balance over time using standard Python
cumulative_winnings_python = [sum(outcomes_python[: i + 1]) for i in range(num_rounds)]
balance_over_time_python = [initial_balance + win for win in cumulative_winnings_python]

# Calculate wins, losses, minimum, and maximum balance using standard Python
wins_python = sum(1 for outcome in outcomes_python if outcome == bet_amount)
losses_python = sum(1 for outcome in outcomes_python if outcome == -bet_amount)
max_balance_python = max(balance_over_time_python)
min_balance_python = min(balance_over_time_python)

# End timing the standard Python implementation
end_time_python = time.time()
time_python = end_time_python - start_time_python

print("Standard Python Implementation:")
print("Wins:", wins_python)
print("Losses:", losses_python)
print("Max Balance:", max_balance_python)
print("Min Balance:", min_balance_python)
print("Execution Time:", time_python, "seconds")
```

# Why NumPy is Faster Than Standard Python

The comparison between standard Python and NumPy implementations of a gambling simulation reveals significant speed differences. Here's why NumPy often outperforms standard Python:

## Vectorization

- NumPy operations are **vectorized**, meaning they're formulated to operate on whole arrays at once. This approach eliminates the need for explicit iteration, leading to substantial speed improvements.

## C Extensions

- NumPy is built in C, allowing it to perform operations much faster than Python's interpreter with Python code. By bypassing the overhead of Python interpreter's dynamic types, NumPy achieves greater efficiency.

## Memory Efficiency

- Unlike Python lists, which are arrays of pointers to objects scattered across memory, NumPy arrays are stored **contiguously** in memory. This storage model allows NumPy to optimize memory access patterns, contributing to its speed.

## Built-in Operations

- Functions like `np.random.choice` and `np.cumsum` are optimized C functions capable of operating directly on NumPy arrays. These functions avoid the type-checking overhead associated with Python lists and loops, further enhancing performance.

## Practical Implications

- For large datasets or computations requiring repetition, NumPy's efficiency gains can be substantial.
- NumPy's optimized and concise syntax not only improves performance but also enhances code readability and maintainability, especially for numerical data operations.

These advantages make NumPy an essential tool for numerical computations in Python, particularly in scenarios where performance is critical.

# Leveraging `.apply()` with Lambda Functions in Pandas

Pandas' `.apply()` method is a powerful tool for applying a function along an axis of the DataFrame or on Series objects. Combined with lambda functions, `.apply()` enables concise and efficient data manipulation.

## `.apply()` Method

- `.apply()` allows you to apply a function across the DataFrame's rows or columns.
- Can be used on both Series and DataFrame objects.
- Especially useful for complex operations that are not covered by built-in methods.

## Lambda Functions

- Lambda functions are small, anonymous functions defined with the `lambda` keyword.
- Ideal for simple operations that can be expressed in a single expression.
- Enhances readability and conciseness when used with `.apply()`.

## Applying Lambda Functions with `.apply()`

- Transform data efficiently.
- Perform row-wise or column-wise operations.
- Can be used for conditional operations, transformations, and more.


```python
import pandas as pd

# Creating a DataFrame
df = pd.DataFrame({"A": range(1, 5), "B": range(10, 50, 10)})
print(df)
# Adding 10 to each element in column 'A'
df["A_plus_10"] = df["A"].apply(lambda x: x + 10)
print(df)
# print("Adding 10 to each element in column 'A':\n", df)
```

```python
# Classifying values in 'A' as 'High' or 'Low'
df["A_class"] = df["A"].apply(lambda x: "High" if x > 2 else "Low")
print("\nClassifying values in 'A':\n", df)
```

```python
# Subtracting 'B' from 'A' and determining if the result is positive
df["A_minus_B_positive"] = df.apply(
    lambda row: "Yes" if row["A"] - row["B"] > 0 else "No", axis=1
)
print("\nSubtracting 'B' from 'A' and checking positivity:\n", df)
```

```python
# Multiplying each element by 2 using lambda function on the entire DataFrame (0 axis)
df_multiplied = df.apply(
    lambda column: column * 2 if column.name in ["A", "A_plus_10"] else column
)
print("\nMultiplying each element by 2 in 'A' and 'B':\n", df_multiplied)
```

### Strategies for Missing Data

- **Keep**: Retain as-is, which may complicate subsequent analyses.
- **Remove**: Drop using `dropna`, potentially losing valuable information.
- **Replace**: Fill in with `fillna`, which can introduce bias if not done carefully.

### Pandas Functions for Missing Data

- `isnull` or `notnull` to detect missing values.
- `dropna` to remove missing data, with `thresh` to specify a threshold.
- `fillna` to impute missing values.


```python
import pandas as pd

# Load the movie scores DataFrame
df_movies = pd.read_csv("movie_scores.csv")

# Detecting missing values
missing_data = df_movies.isnull()

# Output the result
print("Missing Data Detection:\n", missing_data)
```

```python
# Removing rows with missing values
df_no_missing = df_movies.dropna()

# Output the DataFrame with no missing data
print("DataFrame with No Missing Data:\n", df_no_missing)
```

```python
# Filling missing values with a specified value
df_filled = df_movies.fillna(value={"pre_movie_score": 5, "post_movie_score": 5})

# Output the DataFrame with filled missing data
print("DataFrame with Filled Missing Data:\n", df_filled)
```

## GroupBy and Merge Operations in Pandas

Pandas provides SQL-like functionality for grouping and merging datasets, offering powerful data analysis capabilities.

### GroupBy

- Aggregates data around one or more keys.
- Useful for summarizing data and performing group-specific analyses.

### Merge

- Combines two datasets using common keys.
- Similar to SQL joins, with various options (`left`, `right`, `inner`, `outer`).


```python
df = pd.read_csv("tips.csv")

# General groupby
grouped_by = df.groupby("day")

# Group the days together and get mean total_bill per day
grouped = df.groupby("day")["total_bill"].mean()

# Aggregating using different functions
aggregated = df.groupby("day").agg({"total_bill": "mean", "tip": "max"})

print(grouped)
print(aggregated)
```

```python
# Creating two DataFrames to merge
df1 = pd.DataFrame({"key": ["A", "B", "C"], "val": [1, 2, 3]})
df2 = pd.DataFrame({"key": ["B", "C", "D"], "val2": [4, 5, 6]})

# Merging DataFrames
merged_df = pd.merge(df1, df2, on="key", how="right")

print(merged_df)
```

```python
# Creating two DataFrames to merge
df1 = pd.DataFrame({"key": ["A", "B", "C"], "value": [1, 2, 3]})
df2 = pd.DataFrame({"key": ["B", "C", "D"], "value": [4, 5, 6]})

# Merging DataFrames
merged_df = pd.merge(df1, df2, on="key", how="inner", suffixes=("_left", "_right"))

print(merged_df)
```

```python
# Assume df1 and df2 are pre-defined DataFrames for this example

# Inner Merge
inner_merged = pd.merge(df1, df2, on="key", how="inner")

# Output the Inner Merged DataFrame
print("Inner Merged DataFrame:\n", inner_merged)
```

```python
# Outer Merge
outer_merged = pd.merge(df1, df2, on="key", how="outer", suffixes=("_left", "_right"))

# Output the Outer Merged DataFrame
print("Outer Merged DataFrame:\n", outer_merged)
```

```python
# Left Merge
left_merged = pd.merge(df1, df2, on="key", how="left")

# Output the Left Merged DataFrame
print("Left Merged DataFrame:\n", left_merged)
```

```python
# Right Merge
right_merged = pd.merge(df1, df2, on="key", how="right")

# Output the Right Merged DataFrame
print("Right Merged DataFrame:\n", right_merged)
```

```python
# Lets find the correlation between happiness score & GDP per capita
df_without_gdp = pd.read_csv("world_happiness.csv")
df_with_gdp = pd.read_csv("gdp_per_capita.csv")

merged_df = pd.merge(df_without_gdp, df_with_gdp, on="Country or region", how="inner")

correlation = merged_df["Score"].corr(merged_df["GDP per capita"])

print(correlation)
```
